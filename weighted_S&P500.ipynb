```python
import numpy as np #The Numpy numerical computing library
import pandas as pd #The Pandas data science library
import requests #The requests library for HTTP requests in Python
import math #The Python math module
import datetime as dt
import pandas_datareader as web
import fix_yahoo_finance as yf
import requests
from bs4 import BeautifulSoup
from datetime import date
```

```python
#web scraping all symbols in the S&P 500
URL = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find("table", {'id':"constituents"})
df=pd.read_html(str(results))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
stocks = df.sort_values(by=['Symbol']).reset_index(drop=True)
stocks['Symbol'] = df['Symbol'].str.replace('.','-')
stocks = stocks[['Symbol']]
stocks
```

```python
next = stocks.tail(-363)
next.reset_index(drop=True)
```

```python
my_columns = ['Symbol', 'Price','Market_Capitalization', 'Number Of Shares to Buy']
final_dataframe = pd.DataFrame(columns = my_columns)
next_dataframe = pd.DataFrame(columns = my_columns)
final_dataframe
```

```python
next_dataframe = pd.DataFrame(columns = my_columns)
```

```python
def chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python
symbol_groups = list(chunks(next['Symbol'], 100))
symbol_strings = []
for i in range(0, len(symbol_groups)):
    symbol_strings.append(','.join(symbol_groups[i]))
#     print(symbol_strings[i])

next_dataframe = pd.DataFrame(columns = my_columns)

for symbol_string in symbol_strings:
#     print(symbol_strings)
    for symbol in symbol_string.split(','):
        data = web.get_quote_yahoo(symbol)
        next_dataframe = next_dataframe.append(
                                        pd.Series([symbol, 
                                                   data['price'], 
                                                   data['marketCap'],
                                                   'N/A'], 
                                                  index = my_columns), 
                                        ignore_index = True)
        
    
next_dataframe
```
